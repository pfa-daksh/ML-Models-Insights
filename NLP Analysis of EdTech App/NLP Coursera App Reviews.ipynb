```python
from nltk.corpus import stopwords
stopword = stopwords.words("english") # Importing the list of stop words from nltk
```

```python
stopwords_removed = [word for word in list_word if word not in stopword]
print(stopwords_removed) # Removing the stop words from tokenized words from the reviews
```

```python
# Removing stopwords from the list of tokenized sentences from the reviews
import gensim
from gensim.parsing.preprocessing import remove_stopwords
list_sent2 = ""
for sentence in list_sent:
    list_sent2 += remove_stopwords(sentence) 
```

```python
list_sent3 = sent_tokenize(list_sent2)# Now we have sentences with stop words removed
print(list_sent3)
```

```python
from nltk.stem import WordNetLemmatizer
lemma = WordNetLemmatizer()
str4 = ""
for x in range(len(stopwords_removed)):
    stopwords_removed[x] = lemma.lemmatize(stopwords_removed[x])

print(stopwords_removed)# Now we have lemmatized word from the reviews with stop words removed.   
```

```python
from nltk import pos_tag
tagged_wrds = pos_tag(stopwords_removed)
print(tagged_wrds)
```
